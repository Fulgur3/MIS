# Лабораторна робота №4
## Варіант №2
## Білокриницький Олександр, ІПС-41

\begin{cases}
6.2 x_1 + 6.6 x_2 = 83 \\
-9.6 x_1 + 13.8 x_2 = -72 \\
-13.2 x_1 + 5.7 x_2 = 305
\end{cases}

```python
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('seaborn-v0_8-darkgrid')
```

```python
def equation1(x1):
    return (83 - 6.2*x1) / 6.6

def equation2(x1):
    return (-72 + 9.6*x1) / 13.8

def equation3(x1):
    return (305 + 13.2*x1) / 5.7

x1_values = np.linspace(0, 15, 100)
```

```python
plt.figure(figsize=(15,5), dpi=100)

plt.subplot(1,3,1)
plt.plot(x1_values, equation1(x1_values),  color='IndianRed',  linestyle='--')
plt.title('Equation 1: 6.2x1 + 6.6x2 = 83')
plt.subplot(1,3,2)
plt.plot(x1_values, equation2(x1_values) ,color='Teal', linestyle=':')
plt.title('Equation 2: -9.6x1 +  13.8x2 = -72')
plt.subplot(1,3,3)
plt.plot(x1_values,  equation3(x1_values), color='Fuchsia',  linestyle='-.')
plt.title('Equation 3:  -13.2x1 - 5.7x2 = 305')
plt.tight_layout()
plt.show()
```

```python
x1_values = np.linspace(-60, 11, 100)
AB_pairs = [
    (np.array([[6.2, 6.6], [-9.6, 13.8]]), np.array([83, -72])),
    (np.array([[6.2, 6.6], [-13.2, 5.7]]), np.array([83, 305])),
    (np.array([[-9.6, 13.8], [-13.2, 5.7]]), np.array([-72, 305]))
]
solutions = [np.linalg.solve(A, B) for A, B in AB_pairs]
plt.figure(figsize=(10,5), dpi=200)
plt.plot(x1_values, equation1(x1_values), label='6.2x1 + 6.6x2 = 83', color='IndianRed',  linestyle='--')
plt.text(6, equation1(6)-0.5, r'$6.2x_1 + 6.6x_2 = 83$', rotation=-14, color='IndianRed')
plt.plot(x1_values, equation2(x1_values), label='-9.6.2x1 + 13.8x2 = -72', color='Teal', linestyle=':')
plt.text(4, equation2(4)-0.5, r'$-9.6x_1 + -13.8x_2 = -72$', rotation=-10, color='Teal')
plt.plot(x1_values,  equation3(x1_values), label='-13.2x1 + 5.7x2 = 305', color='Fuchsia',  linestyle='-.')
plt.text(8, equation3(8)-1.7,  r'$-13.2x_1 + 5.7x_2 = 305$', rotation=-26, color='Fuchsia')
# Вивід точок перетину
for s in solutions:
   plt.scatter(s[0], s[1], color='Navy', marker='o', zorder=5)
   plt.annotate(f'({s[0]:.2f}, {s[1]:.2f})', (s[0], s[1]), textcoords="offset points", xytext=(-10,5), ha='center', fontsize=10, color='Navy')


plt.fill_between(x1_values, np.minimum(np.maximum(equation2(x1_values), equation1(x1_values)), equation3(x1_values)), np.minimum(np.minimum(equation2(x1_values), equation1(x1_values)), equation3(x1_values)), color='gray', alpha=0.5)
plt.xlabel(r'$x_1$', fontsize=12)
plt.ylabel(r'$x_2$', fontsize=12)
plt.title('Intersection of three lines', fontsize=15)
plt.grid(color = 'purple', linestyle = '--', linewidth = 0.5)
plt.legend()
plt.savefig('graph.jpg')
plt.savefig('graph.png')
plt.savefig('graph.svg')
plt.show()
```
